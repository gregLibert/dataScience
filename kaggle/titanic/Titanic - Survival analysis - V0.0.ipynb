Let's define a plot function that will choose automatically what representation to used based on data type (discrete/continuous) and data structure (number of subgroups)

````python
import numpy as np;
import matplotlib.pyplot as plt;
from matplotlib import cm;
from cycler import cycler

def plotFeatures(name, data, axe):

    if name in DISCRETE_VALUES:
        #feature is discrete
        _uniqueLength = len(data[name].unique())
        # if there is less than 5 groups in dataset => plot a pie
        if _uniqueLength < 5:
            data = data.groupby(name).count()
            data = data[[PRIMARY_KEY]]
            data.columns = ['Count']
            _explode= (data == max(data.Count) )* 0.05
            axe.pie(data, labels=data.index.values,explode=tuple(_explode.values.flatten()), autopct='%1.1f%%', shadow=True, startangle=90)
            axe.axis('equal')
        else:
             # if there is more than 50 groups in dataset => give up
            if _uniqueLength > 50:
                raise ValueError('Too many distinct values')
            # otherwise => plot an histogram
            else:
                data = data.groupby(name).count().sort_values(by=PRIMARY_KEY)#.sort([PRIMARY_KEY])
                data = data[[PRIMARY_KEY]]
                data.columns = ['Count']
                axe.barh(np.arange(_uniqueLength), data['Count'].values, 0.35, tick_label=data.index, color='lightblue', align='center')
                axe.set_ylim([-1, _uniqueLength])
                
        axe.set_xlabel('Count')
    else:
        #feature is of continuous type
        data = data[[name]]
        data = data.dropna()
        axe.violinplot(data.values, showmeans=False, showmedians=True)
        axe.axes.get_xaxis().set_visible(False)

    axe.set_title(name)
    
def plotFeaturesAgainstClasses(name, data, axe):

    _classes = data[CLASS].unique()
    
    if name in DISCRETE_VALUES:
        binWidth = 0.35
        _uniqueValues = data[name].unique()
        _cpt = 0;
        for _class in _classes:
            _data = data[(data[CLASS] == _class)].groupby(name).count()
            _data = _data[_data.columns[0]]
            _data.columns = ['Count']
            
            #just to be sure that each values in group are sorted and present as expected
            _newData = []
            for _value in _uniqueValues:
                _newData.append(_data[(_value == _data.index)].values[0]) if _value in _data.index else _newData.append(0)
            _data = np.array(_newData)
            
            axe.barh(np.arange(len(_data)) + _cpt*binWidth, _data, binWidth, color=cm.jet(1.*_cpt/len(_classes)), label=CLASS_LABELS[_class])
            _cpt += 1

        axe.set_yticklabels(_uniqueValues)
        axe.set_yticks(np.arange(len(_uniqueValues)) + binWidth * _cpt *0.5)
        axe.set_prop_cycle(cycler('color', ['c', 'm', 'y', 'k']) + cycler('lw', [1, 2, 3, 4]))
        axe.legend(loc='upper right')
    else:
        #feature is of continuous type
        _cpt = 0;
        for _class in _classes:
            _data = data[(data[CLASS] == _class)][name]
            _data = _data.dropna()
            violin_parts = axe.violinplot(_data.values, showmeans=False, showmedians=True)
            axe.axes.get_xaxis().set_visible(False)
            for pc in violin_parts['bodies']:
                pc.set_facecolor(cm.jet(1.*_cpt/len(_classes)))
                pc.set_edgecolor('black')
            _cpt += 1

    axe.set_title(name)
    
def plotCategory(name, data, vsClass = False):
    
    _nrows=int(len(name)/2) + len(name) % 2 if not vsClass else int((len(trainDataFrame.columns) - 1)/2) + (len(trainDataFrame.columns ) - 1) % 2
    
    fig, axes = plt.subplots(nrows=_nrows, ncols=2)
    _axes = axes.flat
    fig.set_size_inches(14, _nrows * 7)

    _cpt = 0

    for _feature in name:
        try:
            if (vsClass):
                if _feature != CLASS:
                    plotFeaturesAgainstClasses(_feature, data, _axes[_cpt])
                    _cpt += 1
            else:
                plotFeatures(_feature, data, _axes[_cpt])
                _cpt += 1            
        except ValueError as err:
            print("{} - {}".format(_feature, err))

    plt.tight_layout()
    plt.show() 
````

<h1 id="Data-analysis">Data Analysis</h1>
Let's picture each kept categories against classifications ventilation.

````python
plotCategory(trainDataFrame.columns, trainDataFrame, True)
````

It looks like all these features bring information, even if somes are correlated (Title/Age, Sex/Title...). We will then keep them all for now and try to automatically identify what quantity of information is bringed by each feature.<br/>
To do so, we could compare feature:
- Based on variance and then discare features with low value - (VarianceThreshold)[http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.VarianceThreshold.html#sklearn.feature_selection.VarianceThreshold]
- Based on univariate tests and then keep the best - (keeping the n-best)[http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html#sklearn.feature_selection.SelectKBest] - (keeping those being greater than a percentile of the highest scores)[http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectPercentile.html#sklearn.feature_selection.SelectPercentile]
- By recursively testing the best configuration based on a estimator - (RFECV)[http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFECV.html#sklearn.feature_selection.RFECV]
- Using a random forest to calculate features importance and then manually discare features with low importance  - we will use that technique
- Using LinearDiscriminantAnalysis weight vector (coef\_) to identify most relevant features - (LDA)[http://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html]. Unfortunetlly, projection will give us a 1-dimension space as we only have two classes.

Feature importance will be evaluated by computing how much impurity (wrong classification) each feature will decrease in a decision tree. Using a random forest multiply trees and allows to calculate means of such values.<br/>
There is two side effects with impurity feacture selection method - (see)[http://blog.datadive.net/selecting-good-features-part-iii-random-forests/]:
- There is a bias towards prefering variables with more categories (to avoid overfiting a maximum depth for each tree will be defined)
- For correlated features, once one feature is selected to split data, the correlated feature will have less importance (e.g. once Age is used to classify the data, information brings by Title_Master will be less important). This however has no impact on features selection. 

````python
keptFeatures = trainDataFrame.columns

from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split

#5000 is a bit long but if there is not enough trees, features importance order keep changing places
forest = RandomForestClassifier(n_estimators=5000, random_state=0, n_jobs=-1, max_depth=50)

_training, _testing = train_test_split(pd.get_dummies(trainDataFrame[keptFeatures]), test_size=0.3, random_state=42)

_training_groundTruth = _training["Survived"]
del _training["Survived"]

_testing_groundTruth = _testing["Survived"]
del _testing["Survived"]

forest.fit(_training, _training_groundTruth.values)

featLabels = _training.columns

importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

fig = plt.figure()
ax = plt.axes()

fig.set_size_inches(8, 8)

ax.barh(np.arange(len(indices)), importances[indices], color='lightblue', align='center')

ax.set_yticklabels(featLabels[indices])
ax.set_yticks(np.arange(len(indices)))
plt.ylim([-1, len(indices)])
plt.tight_layout()

plt.title('Feature Importances')
    
plt.show()
````

When can see that most of the information stands in few features ( _Fare, Age, Pclass, Sex_female, Sex_male, Title_Mr, SibSp, Parch_). We will then keep those features and discare the others.

````python
keptFeatures = ["Fare", "Age", "Pclass", "Sex_female", "Sex_male", "Title_Mr", "SibSp", "Parch", "Survived"]

X = pd.get_dummies(trainDataFrame)[keptFeatures]

trainSet_X, testSet_X = train_test_split(X, test_size=0.3, random_state=42)

Y = X["Survived"]
del X["Survived"]

trainSet_Y = trainSet_X["Survived"]
del trainSet_X["Survived"]

testSet_Y = testSet_X["Survived"]
del testSet_X["Survived"]
````

<h1 id="Models-testing">Models testing</h1>
Each model test will follow the same process:
<ol>
<li>Find optimal parameters with gridSearchCV based on self training data set</li>
<li>Draw the learning curve to identify overfit or underfit contexts</li>
<li>Calculate score based on self test data set</li>
<li>Evaluate score and confidence interval with cross validation on all learning set (which is self training set and test set)</li>
</ol>

This is not an optimized approach. We should do score and confidence interval evaluation first then keep the best model and finally find best parameters before calculating score on test data set. But as I want to get all models and not the best one, I change this modus operandi.<br/>

The studied models will be :
* SVC
* RandomForest
* Decision tree
* Adaboost decision tree
* K Neighbors
* Extreme Gradient Boosting

First let's define the printLearningCurve which is directly inspired from Python Machine Learning - Sebastian Raschka.

````python
import matplotlib.pyplot as plt
from sklearn.learning_curve import learning_curve
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

def printLearningCurve(xTrain, yTrain, estimator):
    _pipeline = Pipeline([ ('preprocessing', StandardScaler()), ('modelization', estimator)])
    
    train_sizes_abs, train_scores, test_scores = learning_curve(
                                                estimator=_pipeline,
                                                X=xTrain,
                                                y=yTrain,
                                                train_sizes=np.linspace(0.1, 1.0, 10),
                                                cv=10,
                                                n_jobs=1)
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    fig = plt.figure()
    fig.set_size_inches(8, 8)
    plt.plot(train_sizes_abs, train_mean, color='blue', marker='o', markersize=5, label='training accuracy')
    plt.fill_between(train_sizes_abs, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
    plt.plot(train_sizes_abs, test_mean, color='green', linestyle='--', marker='s', markersize=5, label='validation accuracy')
    plt.fill_between(train_sizes_abs, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
    plt.grid()
    plt.xlabel('Number of training samples')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    #plt.ylim([0.8, 1.0])
    plt.show()
````

##SVM
Parameters:
- C - Outliner penalty.
- Kernel - project data in another space to facilitate segregassion.
- Gamma - kernel attribute

Pipeline is used to add a preprocessing step that normalize (01) input data. gridSearchCV is used to find best parameters in the given range.
> /!\ n_jobs != 1 can silently failed on python when used in a notebook. It seems to be due to how multiprocessing is handled...

````python
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

_pipeline = Pipeline([ ('preprocessing', StandardScaler()), ('modelization', SVC(probability=True, random_state=42))])
_paramRange_penalty = [ 100, 110, 120 ]
_paramRange_gamma = [ 0.001, 0.01, 0.1 ]
_paramGrid = [
    {'modelization__C': _paramRange_penalty, 'modelization__kernel': ['linear']},
    {'modelization__C': _paramRange_penalty, 'modelization__gamma': _paramRange_gamma, 'modelization__kernel': ['rbf']},
    {'modelization__C': _paramRange_penalty, 'modelization__gamma': _paramRange_gamma, 'modelization__kernel': ['sigmoid']}
]

#/!\ n_jobs != 1 silently failed on python2 when used with cross_val_score
gridSearch_SVC = GridSearchCV(estimator=_pipeline, param_grid=_paramGrid, scoring='accuracy', cv=5, n_jobs=-1)

gridSearch_SVC = gridSearch_SVC.fit(trainSet_X, trainSet_Y)
print("Best score: {} with parameters: {}".format(gridSearch_SVC.best_score_, gridSearch_SVC.best_params_))

bestModel_SVC = gridSearch_SVC.best_estimator_
````

Now we will have a look at training and validation accuracies.<br/>
__Good bias/variance trade-off__ is when training and validation accuracies finish with the same "good" accuracy.<br/>
__Overfitting__ or _high variance_ is when training accuracy is far better than validation accuracy.<br/>
__Underfitting__ or _high bias_ is when training and validation accuracies finish both with a low accuracy or when validation accuracy is better than training accuracy (e.g. when dropout is used in neural network).

````python
printLearningCurve(trainSet_X, trainSet_Y, bestModel_SVC)
print('Test accuracy: %.3f' % bestModel_SVC.score(testSet_X, testSet_Y))
````

No obvious overfitting, model seems to stop learning with ~450 samples.<br/>Let's see what can be expect from this type of model with unknown data:

````python
from sklearn.cross_validation import cross_val_score

scores_SVC = cross_val_score(gridSearch_SVC, X, Y, scoring='accuracy', cv=5)

print("CV accuracy: %0.2f (+/- %0.2f)" % (scores_SVC.mean(), scores_SVC.std() * 2))
````

##AdaBoost decision tree

- n_estimators - The maximum number of estimators at which boosting is terminated

GridSearchCV is a little bit different, no pipeline but two classifier linked together. paramGrid keys are updated to match this new configuration.

````python
from sklearn.grid_search import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

_paramRange_max_features = [4, 5, 6 ]
_paramRange_max_depth = [8, 9, 10]
_paramRange_min_samples_leaf = [4, 5, 7, 8]
_paramRange_n_estimators = [ 50, 75, 100 ]
_paramRange_learning_rate = [ 0.5, 0.75, 1.0 ]

_paramGrid = [
    {'base_estimator__max_features': _paramRange_max_features, 
     'base_estimator__max_depth': _paramRange_max_depth, 
     'base_estimator__min_samples_leaf':_paramRange_min_samples_leaf,
     'n_estimators': _paramRange_n_estimators, 
     'learning_rate': _paramRange_learning_rate
     }
]

_dt = DecisionTreeClassifier(random_state=42)

ADT = AdaBoostClassifier(base_estimator = _dt, random_state=42)


gridSearch_ADT = GridSearchCV(estimator=ADT, param_grid=_paramGrid, scoring='accuracy', cv=5, n_jobs=-1)

gridSearch_ADT = gridSearch_ADT.fit(trainSet_X, trainSet_Y)

print("Best score: {} with parameters: {}".format(gridSearch_ADT.best_score_, gridSearch_ADT.best_params_))
bestModel_ADT = gridSearch_ADT.best_estimator_
````

````python
printLearningCurve(trainSet_X, trainSet_Y, bestModel_ADT)
print('Test accuracy: %.3f' % bestModel_ADT.score(testSet_X, testSet_Y))
````
